# 1. Lý thuyết mạng LSTM

## 1.1 Mạng hồi quy RNN

Con người không bắt đầu suy nghĩ của họ từ đầu tại tất cả các thời điểm. Cũng như bạn đang đọc bài viết này, bạn hiểu mỗi chữ ở đây dựa vào từ bạn đã hiểu các chữ trước đó chứ không phải là đọc tới đâu ném hết đi tới đó, rồi lại bắt đầu suy nghĩ lại từ đầu tới chữ bạn đang đọc. Tức là tư duy đã có một bộ nhớ để lưu lại những gì diễn ra trước đó.

Tuy nhiên các mô hình mạng nơ-ron truyền thống thì không thể làm được việc đó, đó có thể coi là một khuyết điểm chính của mạng nơ-ron truyền thống. Ví dụ, bạn muốn phân loại các bối cảnh xảy ra ở tất cả các thời điểm trong một bộ phim, thì đúng là không rõ làm thế nào để có thể hiểu được một tình huống trong phim mà lại phụ thuộc vào các tình huống trước đó nếusử dụng các mạng nơ-ron truyền thống.

Mạng nơ-ron hồi quy (Recurrent Neural Network) sinh ra để giải quyết vấn đề đó. Mạng này chứa các vòng lặp bên trong cho phép thông tin có thể lưu lại được.

<img src="./images/RNN-rolled.png" alt="" width="30%" height="30%"/>

Hình vẽ trên mô tả một đoạn của mạng nơ-ron hồi quy $A$ với đầu vào $x_t$ và đầu ra $h_t$. Một vòng lặp cho phép thông tin có thể được truyền từ bước này qua bước này qua bước khác của mạng nơ-ron.

Các vòng lặp này khiến cho mạng nơ-ron hồi quy trông có vẻ khó hiểu. Tuy nhiên, nếu bạn để ý một chút thì nó không khác mấy so với các mạng nơ-ron thuần. Một mạng nơ-ron hồi quy có thể được coi là nhiều bản sao chép của cùng một mạng, trong đó mỗi đầu ra của mạng này là đầu vào của một mạng sao chép khác. Nói thì hơi khó hiểu, nhưng bạn hãy xem hình mô tả sau:

![RNN unrolled](./images/RNN-unrolled.png)

Chuỗi lặp lại các mạng này chính là phân giải của mạng nơ-ron hồi quy, các vòng lặp khiến chúng tạo thành một chuỗi danh sách các mạng sao chép nhau. Bạn có thấy nó khác gì một mạng nơ-ron thuần không? Không khác gì phải không? Các nút của mạng vẫn nhận đầu vào và có đầu ra hệt như mạng nơ-ron thuần.

Trong vài năm gần đây, việc ứng dụng RNN đã đưa ra được nhiều kết quả không thể tin nổi trong nhiều lĩnh vực: nhận dạng giọng nói, mô hình hóa ngôn ngữ, dịch máy, mô tả ảnh,… Danh sách vẫn còn đang được mở rộng tiếp. Đằng sau sự thành công này chính là sự đóng góp của LSTM. LSTM là một dạng đặc biệt của mạng nơ-ron hồi quy, với nhiều bài toán thì nó tốt hơn mạng hồi quy thuần. Hầu hết các kết quả thú vị thu được từ mạng RNN là được sử dụng với LSTM. Trong bài viết này, ta sẽ cùng khám phá xem mạng LSTM là cái gì nhé.
## 1.2 Vấn đề phụ thuộc xa
Một điểm nổi bật của RNN chính là ý tưởng kết nối các thông tin phía trước để dự đoán cho hiện tại. Việc này tương tự như ta sử dụng các cảnh trước của bộ phim để hiểu được cảnh hiện thời. Nếu mà RNN có thể làm được việc đó thì chúng sẽ cực kì hữu dụng, tuy nhiên liệu chúng có thể làm được không? Câu trả lời là còn tùy.

Đôi lúc ta chỉ cần xem lại thông tin vừa có thôi là đủ để biết được tình huống hiện tại. Ví dụ, ta có câu: “các đám may trên bầu trời” thì ta chỉ cần đọc tới “các đám may trên bầu” là đủ biết được chữ tiếp theo là “trời” rồi. Trong tình huống này, khoảng cách tới thông tin có được cần để dự đoán là nhỏ, nên RNN hoàn toàn có thể học được.

![RNN-shorttermdepdencies](./images/RNN-shorttermdepdencies.png)

Nhưng trong nhiều tình huống ta buộc phải sử dụng nhiều ngữ cảnh hơn để suy luận. Ví dụ, dự đoán chữ cuối cùng trong đoạn: “I grew up in France… I speak fluent French.”. Rõ ràng là các thông tin gần (”I speak fluent”) chỉ có phép ta biết được đằng sau nó sẽ là tên của một ngôn ngữ nào đó, còn không thể nào biết được đó là tiếng gì. Muốn biết là tiếng gì, thì ta cần phải có thêm ngữ cảnh “I grew up in France” nữa mới có thể suy luận được. Rõ ràng là khoảng cách thông tin lúc này có thể đã khá xa rồi.

Thật không may là với khoảng cách càng lớn dần thì RNN bắt đầu không thể nhớ và học được nữa.

Quan sát về nhược điểm của RNN, ta nhận thấy kiến trúc này không hề có cơ chế lọc những thông tin không cần thiết. Bộ nhớ của kiến trúc có hạn, nếu lưu tất cả những chi tiết không cần thiết thì sẽ dẫn đến quá tải, từ đó quên những thứ ở xa trong quá khứ. Tương tự như con người, rất dễ hiểu phải không!!!

![RNN-longtermdependencies](./images/RNN-longtermdependencies.png)

Về mặt lý thuyết, rõ ràng là RNN có khả năng xử lý các phụ thuộc xa (long-term dependencies). Chúng ta có thể xem xét và cài đặt các tham số sao cho khéo là có thể giải quyết được vấn đề này. Tuy nhiên, đáng tiếc trong thực tế RNN có vẻ không thể học được các tham số đó. Vấn đề này đã được khám phá khá sâu bởi <a href="http://people.idsia.ch/~juergen/SeppHochreiter1991ThesisAdvisorSchmidhuber.pdf">Hochreiter (1991) [tiếng Đức]</a> và <a href="http://www-dsi.ing.unifi.it/~paolo/ps/tnn-94-gradient.pdf">Bengio, et al. (1994)</a>, trong các bài báo của mình, họ đã tìm được nhưng lý do căn bản để giải thích tại sao RNN không thể học được.

LSTM và GRU ra đời như bản cải tiến của RNN.

![1_yBXV9o5q7L_CvY7quJt3WQ](./images/1_yBXV9o5q7L_CvY7quJt3WQ.png)

Như hình trên, chúng ta có thể thấy trong mỗi module của 2 kiến trúc trên đều có trang bị các cổng (gate). Nhiệm vụ các cổng này là giúp kiến trúc đánh giá được mức độ quan trọng của thông tin, từ đó đưa ra quyết định giữ lại hay bỏ đi. Nhờ cơ chế này, các thông tin quan trọng có thể được truyền đi rất xa và vẫn có tác động đáng kể trong tương lai xa.

Để hiểu rõ tại sao những kiến trúc này có thể làm được điều đó, chúng ta nên dành chút thời gian để hiểu hơn về cơ chế hoạt động của kiến trức RNN nói chung. Đầu tiên, các từ trong câu được biến đổi thành các vector. sau đó RNN sẽ xử lý các chuỗi vector này từng từ một.

![1_AQ52bwW55GsJt6HTxPDuMA](./images/1_AQ52bwW55GsJt6HTxPDuMA.gif)

Trong lúc xử lý, RNN đưa thông tin về trạng thái ẩn (hidden state) được tính toán khi tính toán trong quá khứ ($w_{t-2}, w_{t-1},$ vv) thành đầu vào cho quá trình xử lý ở hiện tại ($w_{t}$). Trạng thái ẩn này mang thông tin của từ hiện tại và các từ trước đó để truyền lại cho các từ tiếp theo. Đây chính là cơ chế giúp RNN xử lý dữ liệu tuần tự: các từ phía trước sẽ ảnh hưởng đầu ra của từ phía sau.

![1_o-Cq5U8-tfa1_ve2Pf3nfg](./images/1_o-Cq5U8-tfa1_ve2Pf3nfg.gif)

Bây giờ hãy nhìn mộ cách chi tiết hơn làm sao RNN tính toán được trạng thái ẩn. Trước hết, vector của từ hiện tại và trạng thái ẩn của từ phía trước được kết hợp lại, có thể bằng phép concatenation. Vector tổng này sau đó sẽ đi qua hàm tanh. Có thể thấy là tất cả thông tin của các từ phía trước được dồn vào trạng thái ẩn, đây rõ ràng là một nút thắt.

![1_WMnFSJHzOloFlJHU6fVN-g](./images/1_WMnFSJHzOloFlJHU6fVN-g.gif)

Hàm kích hoạt tanh được dùng để điều chỉnh dòng thông tin đi qua hệ thống. Mọi giá trị sẽ được chiếu về khoảng (-1, 1). Khi một vector đi qua mạng neuron, chúng trải qua rất nhiều phép tính và trong quá trình đó sẽ có thành phần nào đó trở nên quá to và khiến các thành phần khác trở nên không đáng kể. Hàm tanh lúc này sẽ giúp tinh chỉnh sự chênh lệch này để giúp "các thành phần yếu thế vẫn có tiếng nói trong xã hội"

![1_gFC2bTg3uihp1klknWU0qg](./images/1_gFC2bTg3uihp1klknWU0qg.gif)

Đó là những thứ căn bản về RNN. Nó hoạt động tương đối hiệu quả với các chuỗi ngắn và tiết kiệm tài nguyên, nhất là khi so sánh với các mô hình trong kiến trúc CNN.

# 1.3 Mạng LSTM

Mọi mạng hồi quy đều có dạng là một chuỗi các mô-đun lặp đi lặp lại của mạng nơ-ron. Với mạng RNN chuẩn, các mô-dun này có cấu trúc rất đơn giản, thường là một tầng tanh.

![LSTM3-SimpleRNN](./images/LSTM3-SimpleRNN.png)

LSTM cũng có kiến trúc dạng chuỗi như vậy, nhưng các mô-đun trong nó có cấu trúc khác với mạng RNN chuẩn. Thay vì chỉ có một tầng mạng nơ-ron, chúng có tới 4 tầng tương tác với nhau một cách rất đặc biệt.

![LSTM3-chain](./images/LSTM3-chain.png)

![LSTM2-notation](./images/LSTM2-notation.png)

LTSM có các cổng giúp lọc thông tin. Trạng thái của cell có thể coi là trí nhớ của toàn hệ thống. Muốn được cập nhất vào hệ thống này, các thành phần thông tin phải đi qua các cổng. Vậy có thể thấy cơ chế này linh hoạt hơn khi mà các thành phần thông tin được chọn lọc mới có thể được đưa vào bộ nhớ, nhờ đó các thành phần có ích từ thời điểm rất lâu trong quá khứ vẫn có thể có tác dụng điều chỉnh.

![1_0f8r3Vd-i4ueYND1CUrhMA](./images/1_0f8r3Vd-i4ueYND1CUrhMA.png)

Trong LSTM sử dùng hàm sigmoid, hàm này có tác dụng như hàm tanh, nhưng vùng chiếu của nó chỉ là $(0,1)$ thay vì $(-1,1)$. Nếu phần thông tin nào không quan trọng, ta chiếu nó về 0, nó coi như bị "lãng quên".

![1_rOFozAke2DX5BmsX2ubovw](./images/1_rOFozAke2DX5BmsX2ubovw.gif)



### 1.3.1 Cổng quên

Cổng này quyết định xem thông tin nào trong bộ nhớ hiện tại được giữ và thông tin nào. Đầu vào là $h_{t-1}$ và $x_t$ rồi đưa qua hàm sigmoid sẽ được giá trị $f_t$ trong khoảng [0, 1]. Đầu ra 1 thể hiện việc giữ toàn bộ thông tin, đầu ra 0 thể hiện toàn bộ thông tin sẽ bị bỏ đi.

![1_GjehOa513_BgpDDP6Vkw2Q](./images/1_GjehOa513_BgpDDP6Vkw2Q.gif)

Công thức cho tầng cổng quên:
$$f_t = \sigma \left( W_f.[h_{t-1}, x_t] + b_f \right)$$

### 1.3.2 Cổng vào

Cổng này dùng để cập nhật thông tin mới. Ở đây có xuất hiện 2 hàm sigmoid và hàm tanh. Tác dụng của chúng cũng như trên. Output từ hàm sigmoid sẽ có tác dụng lọc thông tin đã qua xử lý từ output hàm tanh.

![1_TTmYy7Sy8uUXxUXfzmoKbA](./images/1_TTmYy7Sy8uUXxUXfzmoKbA.gif)

Công thức:
$$i_t = \sigma \left( W_i . [h_{t-1}, x_t] + b_i \right)$$
$$\tilde{C}_t = \text{tanh} \left( W_c . [h_{t-1}, x_t] + b_C \right)$$ 

Giờ là lúc cập nhật trạng thái tế bào cũ $C_{t-1}$ thành trạng thái mới $C_t$. Ta sẽ nhân trạng thái cũ với $f_t$ để bỏ đi những thông tin ta quyết định quên lúc trước. Sau đó cộng thêm $i_t * \tilde{C}_t$. Trạng thái mới thu được này phụ thuộc vào việc ta quyết định cập nhập mỗi giá trị trạng thái ra sao.

![1_S0rXIeO_VoUVOyrYHckUWg](./images/1_S0rXIeO_VoUVOyrYHckUWg.gif)

Công thức:
$$C_t = f_t * C_{t-1} + i_t * \tilde{C}_t $$

### 1.3.3 Cổng ra

Giá trị đầu ra sẽ dựa vào trạng thái tế bào, nhưng sẽ được tiếp tục sàng lọc. Đầu tiên, ta chạy một tầng sigmoid để quyết định phần nào của trạng thái tế bào ta muốn xuất ra. Sau đó, ta đưa nó trạng thái tế bảo qua một hàm tanh để co giá trị nó về khoảng $[-1, 1]$, và nhân nó với đầu ra của cổng sigmoid để được giá trị đầu ra ta mong muốn. Chú ý rằng cả kết quả đầu ra và cả trạng thái cell đều được đưa vào bước tiếp theo.

![1_VOXRGhOShoWWks6ouoDN3Q](./images/1_VOXRGhOShoWWks6ouoDN3Q.gif)

Công thức:
$$o_t = \sigma \left( W_o . [h_{t-1}, x_t] + b_o \right)$$
$$h_t = o_t * \text{tanh}(C_t)$$

# 2. Các biến thể LSTM

Những thứ ta vừa mô tả ở trên là một LSTM khá bình thường. Nhưng không phải tất cả các LTSM đều giống như vậy. Thực tế, các bài báo về LTSM đều sử dụng một phiên bản hơi khác so với mô hình LTSM chuẩn. Sự khác nhau không lớn, nhưng chúng giúp giải quyết phần nào đó trong cấu trúc của LTSM.

Một dạng LTSM phổ biến được giới thiệu bởi <a href="ftp://ftp.idsia.ch/pub/juergen/TimeCount-IJCNN2000.pdf">Gers & Schmidhuber (2000)</a> được thêm các đường kết nối “peephole connections”, làm cho các tầng cổng nhận được giá trị đầu vào là trạng thái tế bào.

![LSTM3-var-peepholes](./images/LSTM3-var-peepholes.png)

Hình trên mô tả các đường được thêm vào mọi cổng, nhưng cũng có những bài báo chỉ thêm cho một vài cổng mà thôi.

Một biến thể khác là nối 2 cổng loại trừ và đầu vào với nhau. Thay vì phân tách các quyết định thông tin loại trừ và thông tin mới thêm vào, ta sẽ quyết định chúng cùng với nhau luôn. Ta chỉ bỏ đi thông tin khi mà ta thay thế nó bằng thông tin mới đưa vào. Ta chỉ đưa thông tin mới vào khi ta bỏ thông tin cũ nào đó đi.

![LSTM3-var-tied](./images/LSTM3-var-tied.png)

Một biến thể khá thú vị khác của LSTM là Gated Recurrent Unit, hay GRU được giới thiệu bởi <a href="http://arxiv.org/pdf/1406.1078v3.pdf">Cho, et al. (2014)</a>. Nó kết hợp các cổng loại trừ và đầu vào thành một cổng “cổng cập nhập” (update gate). Nó cũng hợp trạng thái tế bào và trạng thái ẩn với nhau tạo ra một thay đổi khác. Kết quả là mô hình của ta sẽ đơn giản hơn mô hình LSTM chuẩn và ngày càng trở nên phổ biến.

![LSTM3-var-GRU](./images/LSTM3-var-GRU.png)

GRU chỉ có 2 cổng để chắt lọc thông tin, cổng reset và cổng update. Nó không có trạng thái cell mà chỉ có đầu ra vừa dùng để đưa ra quyết định vừa dùng để thông tin cho các bước tiếp theo.

![1_jhi5uOm9PvZfmxvfaCektw](./images/1_jhi5uOm9PvZfmxvfaCektw.png)

Do mất đi trạng thái cell, nên tác dụng của 2 cổng trên khá khó để phân biệt rõ ràng, có thể nói là cả 2 cổng đều có tác dụng chắt lọc thông tin từ các đầu vào của cell để đưa ra một output thỏa mãn cả 2 tiêu chí là lưu giữ thông tin quá khứ và có khả năng đưa ra quyết định hiện tại một cách chính xác nhất. Do giảm bớt tính phức tạp nên GRU hoạt động nhanh hơn đôi chút so với LSTM. Về tính hiệu quả thì rất khó để đưa ra kết luận chính xác.

Trên đây chỉ là một vài biến thế được chú ý nhiều nhất thôi, thực tế có rất nhiều các biến thể khác nhau của LSTM như Depth Gated RNNs của <a href="http://arxiv.org/pdf/1508.03790v2.pdf">Yao, et al. (2015)</a>. Cũng có những biến thể mà chiến lực xử lý phụ thuộc xa hoàn toàn khác như Clockwork RNNs của <a href="http://arxiv.org/pdf/1402.3511v1.pdf">Koutnik, et al. (2014)</a>.

Nếu bạn muốn tìm hiểu xem biến thể nào là tốt nhất và chúng khác nhau thế nào, thì có thể đọc bài so sánh khá hay này của <a href="http://arxiv.org/pdf/1503.04069.pdf">Greff, et al. (2015)</a>. ## 3.1 LSTM cho dự đoán chuỗi thời gian đơn biến
Chúng ta xem xét data về ngày tháng và lượng khách hàng của một hãng hàng không quốc tế (data `airline-passengers.csv` trong thư mục data cùng jupyter này). Mỗi quan sát được coi là cách nhau 1 tháng nên chúng ta có thể loại trừ cột đầu tiên.

```python
import pandas as pd
import matplotlib.pyplot as plt
dataset = pd.read_csv('./data/airline-passengers.csv', usecols=[1])
dataset
```

```python
plt.plot(dataset)
plt.show()
```

Để áp dụng LSTM cho bài toán này, chúng ta có thể diễn đạt vấn đề như sau. Đó là, với số lượng khách hàng (tính theo đơn vị hàng ngàn) trong tháng này, dự đoán số lượng khách hàng trong tháng tới là bao nhiêu?

Chúng ta có thể viết 1 hàm đơn giản để chuyển đổi cột dữ liệu duy nhất thành tập dữ liệu gồm 2 cột: cột đầu tiên chứa số lượng khách hàng của tháng (t) và cột thứ 2 chứa dữ liệu dự đoán (tháng t+1).

```python
dataset = dataset.values
dataset = dataset.astype('float32')
```

LSTM rất nhạy cảm với độ lớn của dữ liệu đầu vào, đặc biệt khi các hàm sigmoid hay tanh được sử dụng. Nên ta có thể scale dữ liệu của ta lại thành khoảng 0 đến 1, hay còn gọi là chuẩn hoá. Chúng ta dễ dàng chuẩn hoá tập dữ liệu bằng cách sử dụng lớp **MinMaxScaler** từ thư viện scikit-learn.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)
```

Với dữ liệu thời gian, chuỗi giá trị rất quan trọng. Chúng ta sẽ chia dữ liệu 67% dùng để train và 33% dùng để test mô hình

```python
train_size = int(len(dataset)*0.67)
test_size = len(dataset) - train_size
train, test = dataset[0: train_size, :], dataset[train_size:len(dataset), :]
print(len(train), len(test))
```

Bây giờ chúng ta có thể định nghĩa một hàm để tạo một tập dữ liệu mới, như được mô tả ở trên. Hàm lấy hai tham số là **dataset** là mảng numpy mà chsng ta muốn đổi thành tập dữ liệu và **look_back** là số bước thời gian trước sử dụng làm biến đầu vào để dự đoán khoảng thời gian tiếp theo - mặc định là 1.

Mặc định này sẽ tạo ra một tập dữ liệu trong đó X là số lượng hành khách tại thời điểm t và Y là số lượng hành khách tại thời điểm t+1.

```python
import numpy as np
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i+look_back, 0])
    return np.array(dataX), np.array(dataY)
```

```python
look_back = 1
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
```

```python
trainX
```

```python
trainX.shape
```

Mạng LSTM hy vọng dữ liệu đầu vào (X) sẽ được cung cấp với cấu trúc mảng cụ thể dưới dạng: *[sample, timestep, feature ]*

Hiện tại, dữ liệu của tôi ở dạng: *[ sample, featuree ]* và tôi đang đóng khung vấn đề dưới dạng một bước cho mỗi mẫu. Chúng ta có thể chuyển đổi dữ liệu đầu vào và kiểm tra thành cấu trúc dự kiến bằng cách sử dụng **numpy.reshape()** như sau:

```python
# reshape input to be [samples, time steps, features]
trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))
```

```python
trainX
```

Bây giờ chúng ta đã sẵn sàng để thiết kế và phù hợp với mạng LSTM của chúng ta cho vấn đề này.

Mạng có một lớp có thể nhìn thấy với 1 đầu vào, một lớp ẩn với 4 khối LSTM hoặc nơ-ron và một lớp đầu ra tạo ra một dự đoán giá trị duy nhất. 

Để đảm bảo kết quả có thể lặp lại (ngẫu nhiên nhưng lần chạy sau vẫn giữ như vậy), chúng ta dùng lệnh sau:

```python
# fix random seed for reproducibility
np.random.seed(7)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
# create and fit the LSTM network
model = Sequential()
model.add(LSTM(4, input_shape=(1, look_back)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.summary()
model.fit(trainX, trainY, epochs=50, batch_size=1, verbose=1)
```

Cuối cùng, chúng ta có thể tạo dự đoán bằng cách sử dụng mô hình cho cả tập dữ liệu thử nghiệm và train để có được một dấu hiệu trực quan về kỹ năng của mô hình.  hiển thị tập dữ liệu gốc màu xanh lam, các dự đoán cho tập dữ liệu huấn luyện có màu xanh lục và các dự đoán về tập dữ liệu kiểm tra không nhìn thấy có màu đỏ.

```python
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)
# invert predictions
trainPredict = scaler.inverse_transform(trainPredict)
testPredict = scaler.inverse_transform(testPredict)
# shift train predictions for plotting
trainPredictPlot = np.empty_like(dataset)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(trainPredict)+look_back, :] = trainPredict
# shift test predictions for plotting
testPredictPlot = np.empty_like(dataset)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(dataset)-1, :] = testPredict
# plot baseline and predictions
plt.plot(scaler.inverse_transform(dataset))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```

Chúng ta có thể thấy rằng mô hình đã làm tuyệt vời để dự đoán phù hợp với cả bộ dữ liệu đào tạo và thử nghiệm.

# 3.2 LSTM cho dự đoán chuỗi thời gian sử dụng phương pháp cửa sổ (Window method)

Chúng ta có thể diễn đạt vấn đề bằng cách sử dụng nhiều bước thời gian gần đây để đưa ra dự đoán cho bước tiếp theo. Đây được gọi là cửa sổ và kích thước cửa sổ là một tham số có thể điều chỉnh. Ví dụ với thời gian hiện tại t, chúng ta muốn dự đoán giá trị ở lần tiếp theo t+1, chúng ta có thể sử dụng thời gian hiện tại t, và có thể sử dụng cả 2 lần bước trước là t-1 và t-2 là các biến đầu vào.

Hàm **created_dataset()** mà chúng ta đã tạo trong phần trước cho phép chúng ta tạo công thức này cho bài toán chuỗi thời gian bằng cách tăng đối số **look_back** từ 1 lên 3.

Chúng ta có thể chạy lại ví dụ trong phần trước với kích thước cửa sổ lớn hơn.

```python
# convert an array of values into a dataset matrix
def create_dataset(dataset, look_back=1):
	dataX, dataY = [], []
	for i in range(len(dataset)-look_back-1):
		a = dataset[i:(i+look_back), 0]
		dataX.append(a)
		dataY.append(dataset[i + look_back, 0])
	return np.array(dataX), np.array(dataY)
# fix random seed for reproducibility
np.random.seed(7)
# load the dataset
dataframe = pd.read_csv('./data/airline-passengers.csv', usecols=[1])
dataset = dataframe.values
dataset = dataset.astype('float32')
# normalize the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)
# split into train and test sets
train_size = int(len(dataset) * 0.67)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]
# reshape into X=t and Y=t+1
look_back = 3
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
# reshape input to be [samples, time steps, features]
trainX = np.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
testX = np.reshape(testX, (testX.shape[0], 1, testX.shape[1]))
# create and fit the LSTM network
model = Sequential()
model.add(LSTM(4, input_shape=(1, look_back)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.summary()
model.fit(trainX, trainY, epochs=50, batch_size=1, verbose=1)
# make predictions
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)
# invert predictions
trainPredict = scaler.inverse_transform(trainPredict)
testPredict = scaler.inverse_transform(testPredict)
# shift train predictions for plotting
trainPredictPlot = np.empty_like(dataset)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(trainPredict)+look_back, :] = trainPredict
# shift test predictions for plotting
testPredictPlot = np.empty_like(dataset)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(dataset)-1, :] = testPredict
# plot baseline and predictions
plt.plot(scaler.inverse_transform(dataset))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```

# 3.3 Deep LSTM
Các mạng LSTM có thể được xếp chồng lên nhau trong Keras giống như cách các loại lớp khác có thể được xếp chồng lên nhau. Một bổ sung cho cấu hình được yêu cầu là một lớp LSTM trước mỗi lớp LSTM tiếp theo phải trả về chuỗi. Điều này có thể được thực hiện bằng cách đặt tham số **return_sequences** trên lớp thành **True**.

```python
model = Sequential()
model.add(LSTM(4, input_shape=(1, look_back), return_sequences=True))
model.add(LSTM(4))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.summary()
model.fit(trainX, trainY, epochs=50, batch_size=1, verbose=1)
```

```python
# make predictions
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)
# invert predictions
trainPredict = scaler.inverse_transform(trainPredict)
testPredict = scaler.inverse_transform(testPredict)
# shift train predictions for plotting
trainPredictPlot = np.empty_like(dataset)
trainPredictPlot[:, :] = np.nan
trainPredictPlot[look_back:len(trainPredict)+look_back, :] = trainPredict
# shift test predictions for plotting
testPredictPlot = np.empty_like(dataset)
testPredictPlot[:, :] = np.nan
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(dataset)-1, :] = testPredict
# plot baseline and predictions
plt.plot(scaler.inverse_transform(dataset))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```
