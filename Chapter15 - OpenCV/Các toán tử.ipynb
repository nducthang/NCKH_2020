## 1.6 Tạo viền cho hình ảnh

Nếu bạn muốn tạo một viền xung quanh ảnh, bạn có thể sử dụng <a href="https://docs.opencv.org/master/d2/de8/group__core__array.html#ga2ac1049c2c3dd25c2b41bffe17658a36">cv.copyMakeBorder()</a>. Nó có nhiều ứng dụng trong tích chập. Hàm này có các tham số sau:
* **src** - hình ảnh input
* **top, bottom, left, right** - số pixel chỉ độ rộng border theo hướng tương ứng
* **borderType** - Xác định flag cho loại viền thêm vào:
    - cv.BORDER_CONSTANT - Thêm một đường viền có màu không đổi.
    - cv.BORDER_REFLECT - Viền phản chiếu các phần tử. Ví dụ: fedcba|abcdefgh|hgfedcb
    - cv.BORDER_REFLECT_101 hoặc cv.BORDER_DEFAULT - Tương tự như trên, nhưng thay đổi 1 chút. Ví dụ: gfedcb|abcdefgh|gfedcba
    - cv.BORDER_REPLICATE - phần tử cuối cùng được sao chép. Ví dụ: aaaaaa | abcdefgh | hhhhhhh
    - cv.BORDER_WRAP - Không thể giải thích, nó sẽ như này: cdefgh|abcdefgh|abcdefg
    
* **value** - Màu của viền nếu border type là cv.BORDER_CONSTANT

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

BLUE = [255, 0, 0]

img = cv.imread('./data/avatar.jpg')

replicate = cv.copyMakeBorder(img, 100, 100, 100, 100, cv.BORDER_REPLICATE)
reflect = cv.copyMakeBorder(img, 100, 100, 100, 100, cv.BORDER_REFLECT)
reflect101 = cv.copyMakeBorder(img, 100, 100, 100, 100, cv.BORDER_REFLECT_101)
wrap = cv.copyMakeBorder(img, 100, 100, 100, 100, cv.BORDER_WRAP)
constant= cv.copyMakeBorder(img, 100, 100, 100, 100, cv.BORDER_CONSTANT,value=BLUE)

plt.subplot(231),plt.imshow(img,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')

plt.show()
```

# 2. Các phép toán số học trên ảnh

## 2.1 Mục tiêu
* Học về các phép toán số học trên ảnh như cộng, trừ, bitwise...
* Học về các hàm: `cv.add()`, `cv.addWeighted()`,...

## 2.2 Cộng 2 ảnh
Bạn có thể cộng2 ảnh sử dụng hàm `cv.add()`, res = img1 + img2. Cả 2 ảnh nên giống nhau về độ sâu và type, hoặc hình ảnh thứ hai có thể chỉ là một giá trị vô hướng.

```python
x = np.uint8([250])
y = np.uint8([10])

print(cv.add(x,y)) # 250+10=260 => 255

print(x+y) # 250 + 10 = 260 % 256 = 4
```

## 2.3 Trộn ảnh
Đây cũng là cộng hình ảnh, nhưng các trọng số khác nhau được đưa ra cho hình ảnh để mang lại cảm giác hòa trộn hoặc trong suốt.  Hình ảnh được thêm vào theo phương trình dưới đây:
$$g(x)=(1-\alpha)f_0(x)+\alpha f_1(x)$$

Bằng cách thay đổi $\alpha$ từ $0 \to 1$, bạn có thể thực hiện chuyển đổi thú vị giữa hình ảnh này sang hình ảnh khác.

Ở đây tôi có hai hình ảnh để trộn với nhau. Hình ảnh đầu tiên được cho trọng số 0,7 và hình ảnh thứ hai được cho 0,3.  `cv.add WEighted()` áp dụng phương trình sau cho hình ảnh:
$$dst = \alpha . img1 + \beta . img2 + \gamma$$

```python
# gamma = 0
img1 = cv.imread('./data/avatar.jpg')
img2 = cv.imread('./data/anh2.jpg')

# kích cỡ 2 ảnh phải bằng nhau
img1 = img1[0:700, 0:700]
img2 = img2[0:700, 0:700]

dst = cv.addWeighted(img1, 0.7, img2, 0.3, 0)

cv.imshow('dst', dst)
cv.waitKey(0)
cv.destroyAllWindows()
```

## 2.4 Toán tử bitwise

Các toán tử gốm AND, OR, NOT, XOR.

```python
# load 2 images
img1 = cv.imread('./data/avatar.jpg')
img2 = cv.imread('./data/anh2.jpg')

# kích cỡ 2 ảnh phải bằng nhau
img1 = img1[0:700, 0:700]
img2 = img2[0:700, 0:700]

# Đặt ảnh 2 trên góc trái ảnh 1
rows, cols, channels = img2.shape
roi = img1[0:rows, 0:cols]

# tạo mặt nạ cho ảnh
img2gray = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
ret, mask = cv.threshold(img2gray, 10, 255, cv.THRESH_BINARY)

# nghịch đảo mặt nạ
mask_inv = cv.bitwise_not(mask)

img1_bg = cv.bitwise_and(roi,roi,mask = mask_inv)

img2_fg = cv.bitwise_and(img2,img2,mask = mask)

dst = cv.add(img1_bg,img2_fg)
img1[0:rows, 0:cols ] = dst

cv.imshow('res',img1)
cv.waitKey(0)
cv.destroyAllWindows()
```

# 3. Kỹ thuật đo lường và cải tiến hiệu suất
## 3.1 Mục tiêu
Trong xử lý ảnh, vì bạn đang xử lý một số lượng lớn thao tác mỗi giây, nên bắt buộc mã của bạn không chỉ cung cấp giải pháp chính xác mà còn cung cấp nó theo cách nhanh nhất. Vì vậy, trong phần này, bạn sẽ học:
* Cách đo hiệu suất code của bạn
* Một số chú ý để cải tiến hiệu suất code
* Bạn sẽ học sử dụng các hàm: `cv.getTickCount`, `cv.getTickFrequency`,...

Ngoài OpenCV, python cũng cung cấp module **time** rất hữu ích để đo thời gian thực thi.
## 3.2 Đo hiệu suất với OpenCV

```python
img1 = cv.imread('./data/avatar.jpg')
e1 = cv.getTickCount() # số chu kỳ xung nhịp sau mỗi sự kiện tham chiếu
# thực thi 1 đoạn code
for i in range(5,49,2):
    img1 = cv.medianBlur(img1,i)
e2 = cv.getTickCount()
# getTickFrequency trả về tần số chu kỳ xung nhịp
t = (e2 - e1)/cv.getTickFrequency()
print(e1, e2)
print(t)
```

## 3.3 Tối ưu mặc định trong OpenCV
Nhiều hàm OpenCV được tối ưu hóa bằng SSE2, AVX, v.v. Nó cũng chứa mã không được tối ưu hóa. Vì vậy, nếu hệ thống của chúng tôi hỗ trợ các tính năng này, chúng tôi nên khai thác chúng (hầu như tất cả các bộ xử lý hiện đại đều hỗ trợ chúng).  Nó được bật theo mặc định trong khi biên dịch.  Vì vậy, OpenCV chạy mã được tối ưu hóa nếu được bật, nếu không, nó sẽ chạy mã không được tối ưu hóa.  Bạn có thể sử dụng `cv.useOptimized()` để kiểm tra xem nó đã được bật/tắt chưa và `cv.setUseOptimized()` để bật/tắt.  Hãy xem một ví dụ đơn giản.

```python
cv.useOptimized()
```
