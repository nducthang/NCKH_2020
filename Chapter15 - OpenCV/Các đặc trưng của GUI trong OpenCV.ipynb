**Note:** Thay vì sử dụng 3 cờ trên, bạn có thể sử dụng các tương ứng các số nguyên 1, 0 và -1.

### 1.2.2 Hiển thị một hình ảnh
Sử dụng hàm `cv.imshow()` để hiển thị một hình ảnh lên cửa sổ. Cửa sổ hiển thị sẽ tự động fit cho vừa với kích thước hình ảnh.

`cv.waitKey()` là một hàm ràng buộc bàn phím. Tham số của nó là thời gian tính bằng mili giây. Hàm sẽ chờ số mini giây được chỉ định cho bất kỳ sự kiện bàn phím nào. **Note:** Bên cạnh các sự kiện bàn phím ràng buộc, hàm này cũng xử lý nhiều sự kiện GUI khác, vì vậy bạn PHẢI sử dụng nó để thực sự hiển thị hình ảnh.

`cv.destroyAllwindows()`: giải phóng tất cả các cửa sổ tạo ra. Nếu bạn muốn huỷ bất kỳ cửa sổ nào thì có thể dùng hàm `cv.destroyWindow()`, trong đó tham số là tên cửa sổ.

# 2. Bắt đầu với Video

## 2.1 Mục tiêu
* Học về cách đọc video, hiển thị video và lưu video.
* Học về cách kết nối video từ một camera và hiển thị nó.
* Bạn sẽ học sử dụng các hàm: `cv.VideoCapture()`, `cv.VideoWrite()`

## 2.2 Kết nối video từ Camera
Thông thường, chúng ta live stream với một camerra. OpenCv cung cấp một phương thức đơn giản để thực hiện điều này. Để kết nối video, bạn cần tạo một đối tượng **VideoCapture**. Đối số có thể là chỉ mục thiết bị hoặc tên tệp video. Chỉ số thiết bị chỉ là số để chỉ định camera nào. 

```python
import numpy as np
import cv2 as cv

cap = cv.VideoCapture(0)
if not cap.isOpened():
    print("Cannot open camera")
    exit()
while True:
    ret, frame = cap.read()
    
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting...")
        break
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    # display
    cv.imshow('frame', frame)
    if cv.waitKey(1) == ord('q'):
        break
# When everything done, release the capture
cap.release()
cv.destroyAllWindows()
```

Phương thức `cap.read()` trả về `True/False`. Nếu frame đọc chính xác, nó sẽ là `True`.Vì vậy, bạn có thể kiểm tra phần cuối của video bằng cách kiểm tra giá trị được trả về này.

Đôi khi, cap có thể không khởi tạo kết nối capture camera.  Trong trường hợp đó, mã này hiển thị một lỗi. Bạn có thể kiểm tra xem nó có được khởi tạo hay không bằng phương thức `cap.isOpened()`. Nếu là `True`, Ok. Trường hợp khác thì mở nó bằng cách sử dụng `cap.open()`.

Bạn có thể truy xuất một số đặc trưng của video sử dụng phương thức `cap.get(propId)`, trong đó proId là một số từ 0 đến 18. Một vài giá trị có thể chỉnh sửa bằng cách sử dụng `cap.set(propId, value)`, trong đó value là giá trị mới bạn muốn.

Ví dụ, tôi có thể kiểm tra frame chiều rộng và chiều cao với `cap.get(cv.CAP_PROP_FRAME_WIDTH)` và `cap.get(CAP_PROP_FRAME_HEIGHT)`. Nó cho giá trị là 640x480. Nhưng tôi muốn sửa nó thành 320x320, thì sử dụng `cap.set(cv.CAP_PROP_FRAME_WIDTH, 320)` và `ret=cap.set(cv.CAP_PROP_FRAME_HEIGHT, 240)`.

## 2.3 Chạy video từ file
Phát video từ tệp cũng giống như kết nối video từ camera, chỉ cần thay đổi chỉ mục camera thành tên tệp video.  Ngoài ra trong khi hiển thị frame, sử dụng thời gian `cv.waitKey()`.  Nếu tham số quá nhỏ, video sẽ rất nhanh và nếu tham số quá cao, video sẽ chậm (Đây là cách bạn có thể hiển thị video ở chế độ chuyển động chậm). 25 mili giây sẽ ổn trong trường hợp bình thường.

```python
import numpy as np
import cv2 as cv

cap = cv.VideoCapture('./data/Untitled.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    
    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting...")
        break
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    
    cv.imshow('frame', gray)
    
    if cv.waitKey(1) == ord('q'):
        break
cap.release()
cv.destroyAllWindows()
```

## 2.4 Lưu video
Lần này chúng ta tạo một đối tượng `VideoWriter`.  Chúng ta nên chỉ định tên tệp đầu ra (ví dụ: output.avi).  Sau đó, chúng ta nên chỉ định mã `FourCC` (chi tiết dưới đây).  Sau đó, số lượng khung hình trên giây (fps) và kích thước khung hình.  Và cái cuối cùng là cờ `isColor`.  Nếu nó là True, thì là frame màu, nếu không thì frame xám.

`FourCC` là mã 4 byte được sử dụng để chỉ định codec video.  Danh sách các mã có sẵn có thể được tìm thấy trong Fourcc.org. Các codec sau hoạt động tốt:
* Trong Fedora: DIVX, XVID, MJPG, X264, WMV1, WMV2.  (XVID thích hợp hơn. MJPG cho video có kích thước cao. X264 cho video có kích thước rất nhỏ)
* Trong Windows: DIVX 
* Trong OSX: MJPG (.mp4), DIVX (.avi), X264 (.mkv).

Mã FounCC được truyền dưới dạng như sau: `cv.VideoWriter_fourcc('M', 'J', 'P', 'G')` hoặc `cv.VideoWrite_fourcc(*'MJPG')` cho MJPG.

```python
import numpy as np
import cv2 as cv

cap = cv.VideoCapture(0)

# define the codec and create VideoWriter object
fourcc = cv.VideoWriter_fourcc(*'DIVX')
out = cv.VideoWriter('./save/output.avi', fourcc, 20.0, (640, 480))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    
    #frame = cv.flip(frame, 0)
    
    # write the flipped frame
    out.write(frame)
    
    cv.imshow('frame', frame)
    if cv.waitKey(1) == ord('q'):
        break

# Release everything if job is finished
cap.release()
out.release()
cv.destroyAllWindows()
```

Trong tất cả các hàm trên, bạn sẽ thấy một số đối số phổ biến như được đưa ra dưới đây:

### 3.2.1 Drawing Line

Để vẽ một dòng, bạn cần đưa vào tọa độ bắt đầu và kết thúc của dòng. Chúng ta sẽ tạo một hình ảnh màu đen và vẽ một đường màu xanh trên đó từ góc trên bên trái sang góc dưới bên phải.

```python
import numpy as np
import cv2 as cv

# Create a black image
img = np.zeros((515, 512, 3), np.uint8)

# draw a diagonal blue line with thickness of 5 px
cv.line(img, (0,0), (511,511), (255,0,0), 5)
cv.imshow("image",img)
cv.waitKey()
cv.destroyAllWindows()
```

# 4. Sử dụng chuột làm cọ vẽ

## 4.1 Mục tiêu:
* Học về các sự kiện chuột trong OpenCV
* Bạn sẽ học về các hàm: `cv.setMouseCallback()`

## 4.2 Demo đơn giản
Ở đây, chúng tôi tạo ra một ứng dụng đơn giản vẽ một vòng tròn trên một hình ảnh bất cứ nơi nào chúng tôi nhấp đúp vào nó.

Đầu tiên chúng ta tạo một hàm callback để chuột được thực thi khi một sự kiện chuột diễn ra.  Sự kiện chuột có thể là bất cứ điều gì liên quan đến chuột như nút trái xuống, nút trái lên, nhấp đúp chuột trái, v.v. Nó cho chúng ta tọa độ (x, y) cho mỗi sự kiện chuột.  Với sự kiện và địa điểm này, chúng tôi có thể làm bất cứ điều gì chúng tôi muốn.  Để liệt kê tất cả các sự kiện có sẵn, hãy chạy đoạn mã sau:

```python
import cv2 as cv
events = [i for i in dir(cv) if 'EVENT' in i]
print(events)
```

Tạo hàm callback chuột có một định dạng cụ thể:

```python
import numpy as np
import cv2 as cv

# mouse callback function
def draw_circle(event, x, y, flags, param):
    if event == cv.EVENT_LBUTTONDBLCLK:
        cv.circle(img, (x,y), 100, (255, 0, 0), -1)

# Create a black image, a window and bind the function to window
img = np.zeros((512, 512, 3), np.uint8)
cv.namedWindow('image')
cv.setMouseCallback('image', draw_circle)

while(1):
    cv.imshow('image', img)
    if cv.waitKey(20) & 0xFF == 27: # click ESC for exit
        break
cv.destroyAllWindows()
```

## 4.3 Demo nâng cao
Trong phần này, chúng ta vẽ hình chữ nhật hoặc hình tròn (tùy thuộc vào chế độ chúng ta chọn) bằng cách kéo chuột như chúng ta làm trong ứng dụng Paint.  Vì vậy, hàm callback chuột của tôi có hai phần, một phần để vẽ hình chữ nhật và phần khác để vẽ các vòng tròn. Ví dụ cụ thể này sẽ thực sự hữu ích trong việc tạo và hiểu một số ứng dụng tương tác như theo dõi đối tượng, phân đoạn hình ảnh, v.v.

```python
import numpy as np
import cv2 as cv

drawing = False # True nếu chuột được nhấn
mode = True # nếu True, vẽ hình chữ nhật.  Nhấn 'm' để chuyển sang đường cong
ix, iy = -1, -1

# mouse callback function
def draw_circle(event, x, y, flags, param):
    global ix, iy, drawing, mode
    
    if event == cv.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y
    
    elif event == cv.EVENT_MOUSEMOVE:
        if drawing == True:
            if mode == True:
                cv.rectangle(img, (ix,iy), (x,y), (0,255,0), -1)
            else:
                cv.circle(img, (x,y), 5, (0,0,255), -1)
    elif event == cv.EVENT_LBUTTONUP:
        drawing = False
        if mode == True:
            cv.rectangle(img, (ix,iy), (x,y), (0,255,0), -1)
        else:
            cv.circle(img, (x,y), 5, (0,0,255), -1)
            
img = np.zeros((512, 512, 3), np.uint8)
cv.namedWindow('image')
cv.setMouseCallback('image', draw_circle)

while(1):
    cv.imshow('image', img)
    k = cv.waitKey(1) & 0xFF
    if k == ord('m'):
        mode = not mode
    elif k == 27:
        break
cv.destroyAllWindows()
```

## 5.1 Mục tiêu:

## 5.2 Code demo
Ở đây chúng tôi sẽ tạo một ứng dụng đơn giản hiển thị màu bạn chỉ định. Bạn có một cửa sổ hiển thị màu sắc trackbar để chỉ định từng màu B, G, R. Bạn trượt thanh trackbar và thay đổi màu cửa sổ tương ứng. Theo mặc định, màu ban đầu sẽ được đặt thành Đen.

Với hàm <a href="https://docs.opencv.org/master/d7/dfc/group__highgui.html#gaf78d2155d30b728fc413803745b67a9b">cv.createTrackbar()</a>, tham số đầu tiên là tên của trackbar, tham số thứ 2 là tên cửa sổ, tham số thứ 3 giá trị mặc định, tham số thứ 4 là giá trị tối đa, và tham số thứ 5 là hàm callback sẽ thực hiện khi mà các giá trị được thay đổi. Hàm callback luôn có một đối số mặc định là vị trí trackbar. Trong trường hợp của chúng ta, hàm không làm gì cả, vì vậy chúng ta chỉ cần bỏ qua.

Một ứng dụng quan trọng khác của trackbar là sử dụng nó như một button hoặc switch.  OpenCV, theo mặc định, không có hàm button. Vì vậy, bạn có thể sử dụng trackbar để có được chức năng như vậy. Trong ứng dụng của chúng tôi, chúng tôi đã tạo một switch trong đó ứng dụng chỉ hoạt động nếu switch ON, nếu không màn hình luôn có màu đen.

```python
import numpy as np
import cv2 as cv

def nothing(x):
    pass

# create a black image, a window
img = np.zeros((300, 512, 3), np.uint8)
cv.namedWindow('image')

# create trackbars for color change
cv.createTrackbar('R', 'image', 0, 255, nothing)
cv.createTrackbar('G', 'image', 0, 255, nothing)
cv.createTrackbar('B', 'image', 0, 255, nothing)

# create switch for ON/OFF functionality
switch = '0: OFF \n 1: ON'
cv.createTrackbar(switch, 'image',0, 1, nothing)

while(1):
    cv.imshow('image', img)
    k = cv.waitKey(1) & 0xFF
    if k == 27:
        break
    
    # get current positions of four trackbars
    r = cv.getTrackbarPos('R', 'image')
    g = cv.getTrackbarPos('G', 'image')
    b = cv.getTrackbarPos('B', 'image')
    s = cv.getTrackbarPos(switch, 'image')
    
    if s== 0:
        img[:] = 0
    else:
        img[:] = [b, g, r]
cv.destroyAllWindows()
```
